```python
# Pulling data from DC's Data Lake
# Data provided by third party vendor contracted by DC's District Department of Transportation
# Importing the necessary libraries 
import sqlalchemy as sa
from sqlalchemy.engine import create_engine
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
```

```python
# Establishing connection with the Data Lake
connect_args={'auth': 'KERBEROS', 'kerberos_service_name': 'impala'}
engine = create_engine('hive://datbdprddt01:21050', connect_args=connect_args)
```

```python
# Pulling user transaction data from 2018
result_proxy_18  = engine.execute("SELECT meternumber, zone, paymentstart, paymentexpiry, paymentamount, paymenttype, transactionsender, transmissionid FROM thelabdb.ddot_parking_meter_payment WHERE EXTRACT(YEAR FROM paymentstart) = 2018") 
rs_18 = result_proxy_18.fetchall()
```

```python
# Data begins at 01/01/2018
# * Issue with accessing meter and zone information before 2021, 
# will remove 2018-2020 if there is no way to make data usable * 
dc_18 = pd.DataFrame(rs_18)
dc_18.columns = result_proxy_18.keys()

dc_18_col_names = dc_18.columns.tolist()
print(dc_18_col_names)
dc_18.head()
```

```python
# Defining the path for the Parquet file
dc_18_file_path = "dc_2018_parking_payments.parquet"

# Writing the DataFrame to a Parquet file
table_18 = pa.Table.from_pandas(dc_18)
pq.write_table(table_18, dc_18_file_path)

print("Data saved to:", dc_18_file_path)
```

```python
# There is a lack of `meter_id` present for all non-credit card or cash parking payments.
# This issue can be remedied by querying information and joining later joining it
# to the queried transaction data trough joins. This zone data is from DDOT-provided
# file `ddot_parking_meter_app_zones.parquet`
```

```python
# Querying relevant columns from `ddot_parking_meter_app_zones.parquet`
result_proxy_zones  = engine.execute("SELECT meter_id, g_zone, g_parkmobi, blk_no, street, address, numberofsp, longitude, latitude FROM thelabdb.ddot_parking_meter_app_zones") 
rs_zones = result_proxy_zones.fetchall()

dc_zones = pd.DataFrame(rs_zones)
dc_zones.columns = result_proxy_zones.keys()

dc_zones
```

```python
# Defining path for the Parquet file
dc_zones_file = "dc_meters_and_zones.parquet"

# Writing the DataFrame to a Parquet file
dc_zones = pa.Table.from_pandas(dc_zones)
pq.write_table(dc_zones, dc_zones_file)
```

```python
# Pulling parking policies to help create occupancy denominator in later analysis
result_proxy_poli  = engine.execute("SELECT meternumber, policy, effectivedate FROM thelabdb.ddot_parking_meter_policy") 
rs_poli = result_proxy_poli.fetchall()
```
